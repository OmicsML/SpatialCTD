```python
import pandas as pd
spot_gene_expression = pd.read_csv("/content/drive/MyDrive/CosMx_benchmark_pro/data/liver/cancer/new/spot_gene_expression.csv")
spot_gene_expression['fov'] = spot_gene_expression['fov'].astype('int')
del spot_gene_expression['Unnamed: 0']
spot_gene_expression
```

```python
spot_ground_truth = pd.read_csv("/content/drive/MyDrive/CosMx_benchmark_pro/data/liver/cancer/new/ground_truth.csv")
del spot_ground_truth['Unnamed: 0']
spot_ground_truth
```

```python
spot_location = pd.read_csv("/content/drive/MyDrive/CosMx_benchmark_pro/data/liver/cancer/new/spot_location.csv")
spot_location['fov'] = spot_location['fov'].astype('int')
spot_location['spot_id'] = spot_location['spot_id'].astype('int')

del spot_location['Unnamed: 0']
spot_location = pd.merge(spot_ground_truth[['fov', 'spot_id']], spot_location, on=['fov', 'spot_id'])

spot_location
```

```python
spot_gene_expression = pd.merge(spot_location[['fov', 'spot_id']], spot_gene_expression, on=['fov', 'spot_id'])
spot_gene_expression
```

```python
spot_ground_truth = pd.merge(spot_location[['fov', 'spot_id']], spot_ground_truth, on=['fov', 'spot_id'])
spot_ground_truth
```
